# 01-01 : Test HTML to Markdown Conversion

```python
from IPython.display import display, Markdown, HTML
from markdownify import markdownify as md
import json
import requests
```

## 1. Configuration

```python
api_url = "http://localhost:5003"

url_list = [
    'https://www.vodacom.co.za/vodacom/shopping/plans/red',
    'https://www.vodacom.co.za/vodacom/terms/competition/red-suite-experience',
    'https://www.vodacom.co.za/vodacom/services/convenience-and-security/calling/data-travel-bundles',
]
```

## 2. Simple Request

```python
def get_html(url:str, api_url:str):
    """
    Use a POST requests to the /html endpoint of the API to get the HTML of the page.
    
    Args:
        url (str): The URL of the page to get the HTML from.
        api_url (str): The URL of the API.
        
    Returns:
        str: The HTML of the page.
    """
    data = {'url': url}
    headers = {'Content-Type': 'application/json'}
    
    return requests.post(api_url + '/html', data=json.dumps(data), headers=headers)

## test the function
response = get_html(url_list[2], api_url)
#display(HTML(response.json()["content"]))
```

```python
# convert to markdown
strip_tags = ['a', 'img']

md_text = md(
    html=response.json()["content"],
    strip=strip_tags)
display(Markdown(md_text))
```
